```python
#using pandas to clean the data
import pandas as pd
```

```python
#reading the data
df= pd.read_excel("Student Questionnaire_ AI Usage Patterns and Perceptions Among Undergraduate Medical and Dental Students (Responses).xlsx")
df
```

```python
# Removing years or yrs from age and converting into integar
df["Age:"]= df["Age:"].astype(str).str.replace(" years", "").str.replace("years", "").str.replace(" yrs", "").str.replace("yrs", "").astype(int)
df.loc[40:88]
```

```python
#checking data types of all columns 
df.dtypes
```

```python
#giving capital letters to city names to get a good count during data analysis
df["City name:"] = df["City name:"].str.capitalize()
df
```

```python
df.loc[40:88]
```

```python
df["College name:"] = df["College name:"].str.title()
df
```

```python
# replacing likert scales with numerical values
likert_scale = {
    'Strongly Agree': 5,
    'Agree': 4,
    'Neutral': 3,
    'Disagree': 2,
    'Strongly Disagree': 1
}
df.iloc[:,16:48] = df.iloc[:,16:48].replace(likert_scale)
df

```

```python
df.iloc[:,:16]
```

```python
# replacing Nan and Nil values with No
df["Others:"]= df["Others:"].astype(str).str.replace("nan", "No").str.replace("Nil", "No")
print(df["Others:"])
```
